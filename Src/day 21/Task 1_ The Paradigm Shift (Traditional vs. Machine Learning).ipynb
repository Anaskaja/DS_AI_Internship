Day 21: Introduction to Machine Learning
Task 1: The Paradigm Shift
In normal programming, we have to write all the rules ourselves. We give the computer the data and the rules, and it calculates the answer. Machine Learning flips this entirely. Instead, we give the computer the data and the correct answers, and the machine figures out the rules on its own. Once it finds those hidden patterns, it can use them to predict answers for brand-new data without us having to write the logic.

The Three Branches of Machine Learning:
1. Supervised Learning: We train the model using data that already has the correct answers (labels) attached to it.
2. Unsupervised Learning: We give the model raw data with no answers or labels. It has to find patterns, clusters, or groupings all on its own.
3. Reinforcement Learning: An AI learns by interacting with an environment. It gets "rewards" for doing things right and "penalties" for making mistakes.
