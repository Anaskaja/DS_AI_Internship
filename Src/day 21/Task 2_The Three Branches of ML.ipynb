## Task 2: The Three Branches of ML

**1. Supervised Learning:** The data has labels or "answers." The model learns from an answer key to make future predictions.

**2. Unsupervised Learning:** The data has no labels. The model must find hidden clusters, patterns, or groupings entirely on its own.

**3. Reinforcement Learning:** An AI learns via a system of rewards and penalties through trial and error in an environment.

**Real-World Examples Comparison Table:**

| ML Category | Real-World Example 1 | Real-World Example 2 |
|---|---|---|
| **Supervised** | Spam Detection: Identifying emails as Spam or Inbox based on past labeled emails. | Medical Diagnosis: Predicting if a tumor is malignant based on historical patient data. |
| **Unsupervised** | Customer Segmentation: Finding hidden clusters of shoppers with similar habits. | Recommendation Systems: Grouping similar movies to recommend to users. |
| **Reinforcement** | Autonomous Vehicles: A self-driving car learning to navigate traffic safely. | Game AI: A computer learning to play chess through millions of wins and losses. |

```python
import pandas as pd

# 1. Simulating a mock dataset 
data = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'AnnualIncome': [50000, 65000, 120000, 80000, 150000],
    'Purchased_Premium': [0, 0, 1, 0, 1] # This is our Target Label (the answer)
})

# 2. Correctly dropping the target column to create Features (X)
X = data.drop("Purchased_Premium", axis=1)

# 3. Isolating the target column to create Labels (y)
y = data["Purchased_Premium"]

print("Features (X) successfully separated (No Data Leakage):")
display(X)

print("\nLabels (y) successfully isolated:")
display(y)
```
